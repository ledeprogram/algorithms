```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('2013_NYC_CD_MedianIncome_Recycle.xlsx')
df
```

```python
df.plot(kind="scatter", x="MdHHIncE", y="RecycleRate")
```

```python
lm = smf.ols(formula="RecycleRate~MdHHIncE",data=df).fit()
lm.params
```

```python
median_income_input = input('Please type down a median income you want to know the recycle rate for:')
if median_income_input:
    output = (float(lm.params['MdHHIncE']) * float(median_income_input)) + (float(lm.params['Intercept']))
    print("The recycle rate of a community district with a median income of", median_income_input, "US-dollar has a recycling rate of approx.", round(output, 2))
```
