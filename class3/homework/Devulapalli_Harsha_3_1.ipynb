```python
#Sorting Algorithm

def sortingalgo(numberslist):
    print("The unsorted list is",numberslist)
    import time
    start_time = time.time()
    sorted_list = []
    while numberslist:
        minimum = numberslist[0] 
        for x in numberslist: 
            if x < minimum:
                minimum = x
        sorted_list.append(minimum)
        numberslist.remove(minimum)    
    print("")
    print("The sorted list is",sorted_list)
    print("")
    print("--- %s seconds ---" % (time.time() - start_time))
```

```python
sortingalgo([3,10,5,2])
```
