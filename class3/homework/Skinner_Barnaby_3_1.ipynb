```python
#Min_sorting, this is cheeky, but works nicely. Probably not fair to use min though
def min_sorting(lst):
    new_list = []
    for number in range(len(lst)):
        lowest_number = min(lst)
        new_list.append(lowest_number)
        lst.remove(lowest_number)
    return new_list
```

```python
#Insertion_sorting, this is what we covered in class
def insertion_sorting(lst):
    for index in range(1,len(lst)):
        currentvalue = lst[index]
        while index > 0 and lst[index-1] > currentvalue:
            lst[index]=lst[index-1]
            index = index - 1
        lst[index] = currentvalue
    return lst
```
