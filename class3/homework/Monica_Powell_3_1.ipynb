have a counter to consider the position and set it to = 0
for each item in the sorted list:
    if the input number is equal to the item in that position, print the counter
    increment the counter
    if not found, print not there

```python
def sort(list):
    position = 0
    sorted_list = []
    original_len = len(list) - 1
    
    while position < original_len:
        print("List length is", original_len)
        current_num = 0 
        min = list[0]
        print(current_num, "current_num", "and lenlist-1", len(list))
        while current_num <= (len(list)):
            
            # print(list[current_num + 1])
            print("min", min)
            if list[current_num] < min:
                min = list[current_num]
                sorted_list.append(min)
            else:
                if current_num < (len(list) - 1):
                    current_num = current_num + 1
       
        list.remove(min)
        position = position + 1
   
        
        
    return sorted_list
```
